In this notebook, we will explore the perceptron's ability to classify data. 



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([1,1])
x = np.array([-1,1])
origin = np.array([0,0]) # origin point

x_coords = np.arange(-2,2,0.01)
y_coords = -x_coords
upper_y_coords = y_coords + 4

plt.axis([-2,2,-2,2])
plt.fill_between(x_coords,y_coords,upper_y_coords,facecolor="orange", # The fill color
                 color='blue',       # The outline color
                 alpha=0.2)

plt.quiver(*origin, *w, color=['b'], angles='xy', scale_units='xy', scale=1.)
plt.quiver(*origin, *x, color=['r'], angles='xy', scale_units='xy', scale=1.)
plt.show()
```

```python
w @ x
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))

ax.quiver(*origin, *w, color=['b'], scale_units = 'x', scale=1.)
ax.axis([-2,2,-2,2]);
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5));

ax.quiver(*origin, *w, color=['b'], scale_units = 'x', scale=1.);
ax.axis([-2,2,-2,2]);
ax.fill_between(x_coords,y_coords,upper_y_coords,facecolor="orange", color='blue', alpha=0.2);
```

```python
x_a = [-1,1]
x_b = [1,-1]

fig, ax = plt.subplots(1,1,figsize=(5,5));
ax.quiver(*origin, *w, color=['b'], scale_units = 'x', scale=1.);
ax.quiver(*origin, *x_a, color=['black'], scale_units = 'x', scale=1.);
ax.quiver(*origin, *x_b, color=['black'], scale_units = 'x', scale=1.);
ax.axis([-2,2,-2,2]);
ax.fill_between(x_coords,y_coords,upper_y_coords,facecolor="orange", color='blue', alpha=0.2);
```
